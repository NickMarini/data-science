```python
import matplotlib.pyplot as plt

# Assuming your DataFrame is named 'df' and it has the following columns:
# '1st HR bpm', 'Change in HR', and 'Category'

# Create a scatter plot and color by category
categories = df['Group'].unique()

# Create the scatter plot
for category in categories:
    subset = df[df['Group'] == category]
    plt.scatter(subset['1st HR bpm'], subset['Change in HR'], label=category)

# Add labels and title
plt.xlabel('1st HR bpm')
plt.ylabel('Change in HR')
plt.title('Scatter Plot of 1st HR bpm vs Change in HR with Categories')

# Add legend
plt.legend(title='Category')

# Show plot
plt.show()
```
